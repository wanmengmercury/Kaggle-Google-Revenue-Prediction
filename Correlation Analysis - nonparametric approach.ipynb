```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

import math
```

```python
X = pd.read_csv('/Users/mercuryliu/Documents/Kaggle/ga-customer-revenue-prediction/X_new.csv').drop('Unnamed: 0', axis=1)
y = pd.read_csv('/Users/mercuryliu/Documents/Kaggle/ga-customer-revenue-prediction/y_new.csv').drop('Unnamed: 0', axis=1)
```

```python
X.columns
```

```python
r_sp = []
p_val =[]
for i in range(len(X.columns)):
    a,b = stats.spearmanr(X.iloc[:,i],y)
    r_sp.append(a)
    p_val.append(b)

```

```python
r_sp_all = pd.DataFrame(r_sp, columns=['speansmansrank'])
```

```python
r_sp_all['p_value'] = p_val
```

```python
t_obs_spear = r_sp[0] * math.sqrt((len(X) - 2) / (1 - r_sp[0]**2))
print(t_obs_spear)
```

```python
#find t alpha/2, n-2, suppose we want 95% confidence level
t = stats.t.ppf(1-0.05/2, len(X)-2)
#H_0: ro = 0 (no association between X[i] and total revenue)
#H_1: ro is not equal to zero 
t_obs = []
for i in range(len(r_sp)):
    t_test =  r_sp[i] * math.sqrt((len(X) - 2) / (1 - r_sp[i]**2))
    if abs(t_test) >= t:
        t_obs.append('reject')
    else: 
        t_obs.append('doesnt reject')
```

```python
#hypothesis testing results
r_sp_all['H.T'] = t_obs
```

```python
r_sp_all['X_names'] = X.columns
```

```python
#we can claim correlation based on the results below
r_sp_all.sort_values(by=['speansmansrank'], ascending=False)
```
